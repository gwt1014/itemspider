## 淘宝的搜索页
会包含天猫的商品，需要把商品连接找然后区分是淘宝还是天猫的。

```python
def getitemurls_taobao(target,sort='sale-desc',maxpage=1):
    itemurls = {}
    itemurl_taobao = []
    itemurl_tmall = []
    for page in range(0,maxpage):
        params = {
                    'q' : target,
                    'sort' : sort,
                    's' : str(44*(page))
        }
        res = requests.get(init_searchurl_taobao,params=params,headers=headers)
        ids = re.findall('"nid":"(.*?)","category"',res.text,re.S)
        urls = re.findall('"detail_url":"//(.*?)","view_price"',res.text,re.S)
#        print(len(ids))
#        print(len(urls))
        if len(ids) == 0:
            print('被干了这次不算')
        info = dict(ids = ids , urls=urls)
        for i in range(0,len(ids)):
        #由于淘宝页面返回的URL是带//OOOC3这种乱码的，所以url只是看一下到底是淘宝还是天猫，并不作为真正的url传回去
        #返回的URL是用init_itemurl拼出来的
            if 'tmall' in info['urls'][i]:
                itemurl_tmall.append(init_itemurl_tmall.format(info['ids'][i]))
            elif 'taobao' in info['urls'][i]:
                itemurl_taobao.append(init_itemurl_taobao.format(info['ids'][i]))  
#        print("稍等一下")
        time.sleep(random.randint(0,1))
    itemurls['tmall'] = itemurl_tmall
    itemurls['taobao'] = itemurl_taobao
    print('共有商品%d个，其中淘宝%d个，天猫%d个'%((len(itemurl_tmall)+len(itemurl_taobao)),len(itemurl_taobao),len(itemurl_tmall)))
    return itemurls
```

# 获取评论

 先做一个总的函数，用来区分需要获得评论的物品是淘宝的还是天猫的，然后分别做两个小函数去爬取评论

```python
def getcomments(itemurls):
# 分别取出淘宝和天猫的商品链接
    urls_taobao = itemurls['taobao']
    urls_tmall = itemurls['tmall']
    taobao = pd.DataFrame
    tmall=pd.DataFrame
#爬淘宝评论
    result_taobao = []
    result_taobao_notnull = []
    for i in range(0,len(urls_taobao)):
        result_taobao.append(getcomments_taobao(urls_taobao[i]))
    for i in result_taobao:
        if len(i) != 0:
            result_taobao_notnull.append(i)
    taobao = pd.concat(result_taobao_notnull,axis=0,ignore_index=True)
#爬天猫评论        
    result_tmall = []
    result_tmall_notnull = []
    for i in range(0,len(urls_tmall)):        
        result_tmall.append(getcomments_tmall(urls_tmall[i]))
    for i in result_tmall:
        if len(i) != 0:
            result_tmall_notnull.append(i)
    tmall = pd.concat(result_tmall_notnull,axis=0,ignore_index=True)
#拼在一起
    comments = taobao.merge(tmall,how='outer')
    return comments
```

## 获取天猫的评论

```python
#获取单一商品评论的函数，内部会调用一个获取商品多页评论的递归函数
# 传入的就是这个商品URL
def getcomments_tmall(itemurl):
    #先拿到商品ID
    itemid = re.findall('id=(\d+)',itemurl)[0]
    #获取总评论数，评论页数
        #先拼出来评论的第一页
    t = str(time.time()*1000).split('.')
    params = {
            'currentPage':1,
             'itemId': itemid,
             '_ksTS': '%s_%s' % (t[0], t[1]),
             'callback': 'jsonp%s' %(int(t[1])+1),
             'content':1,
             'order':3 # order = 3 是默认顺序 order=1是按时间排序
            }
    
    res = requests.get(init_commenturl_tmall,params=params)
    jc = json.loads(re.findall(r'{.*}',res.text)[0])
    if len(jc['rateDetail']) != 0:
        ## 这里也是有可能被淘宝干掉。。。。。
        totcom = jc['rateDetail']['paginator']['items']
        maxpage = min(jc['rateDetail']['paginator']['lastPage'],99) # 最大是99
        print('\r    总评论%d条，可获取%d页(天猫最多99页)'%(totcom,maxpage))
        #有了最大页数，就做一个递归函数把最大页数里面所有的都爬下来
        pagelist = list(range(1,maxpage+1))
    # 设置个info列表，每一页的信息为一个元素
        info = []
        geteachpagecom_tmall(itemid,pagelist,info)
        itemcomments = pd.DataFrame
    #   把info这个list的每个dataframe拼成一个dataframe\
        print("该商品所有页面都爬完了")
        if len(info) != 0:
            itemcomments = pd.concat(info,axis = 0 ,ignore_index=True)
            #itemcomments.to_csv('comments.csv',mode = 'a+',header=False,index=False)
            itemcomments.to_csv('output/'+target+'-comments-'+str(time.strftime("%Y-%m-%d"))+'.csv',mode = 'a+',header=False,index=False)
        else :
            print("没评论")
        return  itemcomments 
```
